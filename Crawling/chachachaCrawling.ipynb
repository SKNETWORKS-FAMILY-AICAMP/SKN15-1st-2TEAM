```python
#차 id 뽑기
#10페이지까지
carseq_list = []

    
for page in range(1, 11):
    url = f"https://www.kbchachacha.com/public/search/main.kbc?page={page}"
    driver.get(url)
    time.sleep(2)

    bs = BeautifulSoup(driver.page_source)

    ids = bs.find_all("a", class_="link")

    for id in ids:
        #href 속성이 없다면 공백 반환
        href = id.get("href", "")
        #숫자 찾는 정규식
        match = re.search(r'carSeq=(\d+)', href)
        if match:
            carseq = match.group(1)
            carseq_list.append(carseq)

driver.close()
```

```python
#전체 페이지 id 뽑기
n=1
car_seqs = []
failed_pages =[]
while 1:
    try:
        driver.get(url.format(n))
        time.sleep(0.5)
        html = driver.page_source
        soup = bs(html, 'html.parser')
        as1 = soup.find_all('div', class_='area')
        for a in as1:
            car_seqs.append(a['data-car-seq'])
        car_seqs = list(set(car_seqs))
        print('페이지 : ', n, 'seq 갯수 : ', len(car_seqs))
        n +=1
    except Exception as e:
        print('오류발생 페이지 :', n)
        failed_pages.append(n)
        n +=1
        continue
```

```python
# 지피티 코드
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

wait = WebDriverWait(driver, 10) 

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

# 차량 ID 목록
car_seqs = ['26599428',
 '26856904',
 '27072232',
 '26466222',
 '25944369',
 '26926381',
 '26906201',
 '26678393',
 '26982924',
 '26695655',
 '27127196',
 '27030333',
 '25336036']

for car_seq in car_seqs:
    url = f"https://www.kbchachacha.com/public/car/detail.kbc?carSeq={car_seq}"
    driver.get(url)
    time.sleep(3)  # 페이지 로딩 대기 (WebDriverWait 사용도 가능)

    try:
        car_model = driver.find_element(By.CSS_SELECTOR, "strong.car-buy-name").text
        car_price = driver.find_element(By.CSS_SELECTOR, "div.car-buy-price dl dd strong").text

        spans = driver.find_elements(By.CSS_SELECTOR, "div.txt-info span")
        car_year = spans[0].text if len(spans) > 0 else "정보 없음"
        car_distance = spans[1].text if len(spans) > 1 else "정보 없음"
        car_fuel_type = spans[2].text if len(spans) > 2 else "정보 없음"

        seller_name = driver.find_element(By.CSS_SELECTOR, "span.name").text
        company_name = driver.find_element(By.CSS_SELECTOR, "span.place-add").text

        map_txt = driver.find_elements(By.CSS_SELECTOR, "div.map-txt p")
        company_branch = map_txt[0].text if len(map_txt) > 0 else "정보 없음"
        company_address = map_txt[1].text if len(map_txt) > 1 else "정보 없음"

        company_tel = driver.find_element(By.CSS_SELECTOR, "div.dealer-tel-num").text

        print(f"[{car_seq}]")
        print("모델:", car_model)
        print("가격:", car_price)
        print("연식:", car_year)
        print("주행거리:", car_distance)
        print("연료:", car_fuel_type)
        print("판매자:", seller_name)
        print("업체명:", company_name)
        print("지점명:", company_branch)
        print("주소:", company_address)
        print("전화번호:", company_tel)
        print("------")

    except Exception as e:
        print(f"[{car_seq}] 오류 발생: {e}")

driver.quit()
```
