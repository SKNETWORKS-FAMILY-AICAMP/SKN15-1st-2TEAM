```python
def page_car(page_num):
    url = f"https://m.bobaedream.co.kr/mycar/mlist/A/{page_num}?gubun=A"
    bobaedream_url = requests.get(url)
    bs = BeautifulSoup(bobaedream_url.text)
    bs = bs.find("ul",class_="car-list").find_all('a',class_="link")
    bs[0] # 짝수엔 키 번호와 사진 / 홀수에는 정보들들
    car_id = []
    for idx,i in enumerate(bs):
        if idx % 2 == 0:
            car_id.append(i['href'].split('/')[-1]) # 차 고유 id값 추출

    return car_id


list_car_id = [] # 차 고유 id값 리스트
for page_number in range(1,72):
    list_car_id.append(page_car(page_number))
```

```python


def Crawling(car_numberid,mode):
# 차 번호,링크
    info_list = []
    info_car_url = f"https://m.bobaedream.co.kr/cyber/cview/{car_numberid}"
    r = requests.get(info_car_url)
    r.encoding = 'utf-8' 
    main_bs = BeautifulSoup(r.text)
    bs = main_bs.find("div",class_="bobae-head sub js-header").find("h1")
    car_number = bs.text 
    info_list.append(car_number)
    info_list.append(info_car_url)

    # 모델명, 연식, 주행거리, 연료, 가격
    bs = main_bs.find('div',"product-summary")
    bs = bs.text.split("\n")
    for car_data in bs:
        if car_data == '':
            continue
        else:
            info_list.append(car_data)

    # 색상, (수입,국산), 차대번호
    bs = main_bs("div",class_="ui-data-list")
    bs = bs[0].text 
    bs = bs.replace("\n"," ").strip().split("     ")

    count = 0
    car_count = 0
    for idx,i in enumerate(bs):
        if i.split()[0] != "수입구분":
            count == 1
        if i.split()[0] != "차대번호":
            car_count == 1
    
    if count != 1:
        info_list.append("국산")

    if car_count != 1:
        info_list.append("null")

    for idx,i in enumerate(bs):
        if i.split()[0] == "색상" or i.split()[0] == "차대번호":
            info_list.append(" ".join(i.strip().split(" ")[1:]))

    # 제원 옵션 (연비만 뽑기)
    bs = main_bs
    bs = bs.find("div",class_="swiper-wrapper").find_all("li")
    for idx,i in enumerate(bs):
        try:
            if idx == 1:
                info_list.append((i.text.strip().split("\n")[1]))
        except:
            info_list.append("없음")

    # 옵션정보
    bs = main_bs
    try:
        bs = bs.find_all("div",class_ = "swiper-wrapper")[1].find_all('li')
        for i in bs:
            if i['data-option'] == 'true':
                info_list.append(i.text.replace("\n",""))
    except:
        pass

    def safe_search(pattern, text, flags=0):
        match = re.search(pattern, text, flags)
        return match.group(1).strip() if match else None

    """
    https://m.bobaedream.co.kr/myinfo/seller/no/{2245469}/t/mycar/
    https://m.bobaedream.co.kr/myinfo/seller/no/{car_numberid}/t/cyber/
    """
    if mode == 1:
        info_car_url = f"https://m.bobaedream.co.kr/myinfo/seller/no/{car_numberid}/t/mycar/" # https://m.bobaedream.co.kr/myinfo/seller/no/{id}/t/cyber/
    else:
        info_car_url = f"https://m.bobaedream.co.kr/myinfo/seller/no/{car_numberid}/t/cyber/" # https://m.bobaedream.co.kr/myinfo/seller/no/{id}/t/cyber/
    
    r = requests.get(info_car_url)
    r.encoding = 'utf-8' 
    main_bs = BeautifulSoup(r.text)
    bs = main_bs.find("div",class_="ui-data-list").text

    result = {
        "이름": safe_search(r'이름\s*\n\s*([^\n]+)', bs)[:3],
        "상사명": safe_search(r'상사명\s*\n\s*([^\n]+)', bs),
        "조합명": safe_search(r'조합명\s*\n\s*([^\n]+)', bs),
        "주소": safe_search(r'주소.*?\n\s*(.+?)\n{2,}', bs, re.DOTALL),
        "가입일": safe_search(r'가입일\s*\n\s*([^\n]+)', bs),
        "사원증": safe_search(r'사원증\s*\n\s*([^\n]+)', bs),
        "판매중": safe_search(r'판매중\s*\n\s*([0-9]+)', bs),
        "판매완료": safe_search(r'판매완료\s*\n\s*([0-9]+)', bs),
    }  

    for key, value in result.items():
        if key == "이름" or key == "상사명" or key == "주소" or key == "사원증":
            if value == "":
                info_list.append("null")
                
            else:    
                info_list.append(f"{value}")

    return info_list
```
